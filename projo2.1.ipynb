```python
from urllib.request import urlopen
import requests
from bs4 import BeautifulSoup as bts
import pandas as pd
import re
import time
import sys
from datetime import datetime
from datetime import date
```

```python
def getAndParseURL(url):
    result = requests.get(url,headers={"User-Agent":"Mozilla/5.0"})
    soup = bts(result.text, 'html.parser')
    return soup
```

```python
#sayfaların toplanması.

PAGE_NUM = []
def find_page_num():
    
    html = getAndParseURL("https://www.imdb.com/search/title/?groups=top_1000&sort=user_rating,desc")
    cntr = html.find("a",{"class":"lister-page-next next-page"})
    for i in range(0,1000,50):
        
        PAGE_NUM.append("https://www.imdb.com/search/title/?groups=top_1000&sort=user_rating,desc"+str("&start=")+str(i))
    return PAGE_NUM
find_page_num()
```

```python
result = []
for pr in ALL_PRODUCT[:20]:
    
    
    prod = getAndParseURL(pr)
    MOVIE_NAME=prod.find("h1",{"data-testid":"hero-title-block__title"}).text
    YEAR=prod.find("span",{"class":"TitleBlockMetaData__ListItemText-sc-12ein40-2 jedhex"}).text
    RUNTIME=prod.find("li",{"data-testid":"title-techspec_runtime"}).text[7:]
    IMDB=prod.find("span",{"class":"AggregateRatingButton__RatingScore-sc-1ll29m0-1 iTLWoV"}).text
    
    metaCheckPoint = prod.find("span",{"class":"score-meta"})
    
    if metaCheckPoint:
        
        METASCORE = prod.find("span",{"class":"score-meta"}).text
    else:
        METASCORE = "-"
    
    GENRE=prod.find("li",{"data-testid":"storyline-genres"}).li.text
    
    budgetCheckPoint = prod.find("li",{"data-testid":"title-boxoffice-budget"})
    
    if budgetCheckPoint:
        
        BUDGET=prod.find("li",{"data-testid":"title-boxoffice-budget"}).text[7:18] 
    else:
        BUDGET = "-"
        
       
    opweCheckPoint = prod.find("li",{"data-testid":"title-boxoffice-openingweekenddomestic"})
    
    if opweCheckPoint:
        
        OPENING_WEEKEND =prod.find("li",{"data-testid":"title-boxoffice-openingweekenddomestic"}).li.span.text
        
    else:
        OPENING_WEEKEND = "-"
        
        
    worldwideCheckPoint =prod.find("li",{"data-testid":"title-boxoffice-cumulativeworldwidegross"})
    
    if worldwideCheckPoint:
        
        WORLDWIDE_GROSS = prod.find("li",{"data-testid":"title-boxoffice-cumulativeworldwidegross"}).li.span.text
        
    else:
        
        WORLDWIDE_GROSS = "-"
        


    result.append([MOVIE_NAME,YEAR,RUNTIME,IMDB,METASCORE,GENRE,BUDGET,OPENING_WEEKEND,WORLDWIDE_GROSS])
    
    
    
column = ["MOVIE_NAME","YEAR","RUNTIME","IMDB","METASCORE","GENRE","BUDGET","OPENING_WEEKEND","WORLDWIDE_GROSS"]

dfMatch = pd.DataFrame.from_records(result, columns=column)
        
```
